```python
S = [1] * 13
H = [1] * 13
C = [1] * 13
D = [1] * 13
n = int(input()) 
for i in range(n):
    s,num = input().split()
    num = int(num)
    if s == "S":
        S[num - 1] = 0
    if s == "H":
        H[num - 1] = 0
    if s == "C":
        C[num - 1] = 0
    if s == "D":
        D[num - 1] = 0
print()
for i in range(13):
    if S[i] == 1:
        print("S " + str(i + 1))
for i in range(13):
    if H[i] == 1:
        print("S " + str(i + 1))
for i in range(13):
    if C[i] == 1:
        print("S " + str(i + 1))
for i in range(13):
    if D[i] == 1:
        print("S " + str(i + 1))
```
